```python
import pandas as pd
from evaluation import top_bottom
from visualisation import plot_score_hist, plot_year_mean
import json
```

```python
def return_df(MODEL_NAME):
    df = pd.read_csv(f"../data/processed/paper_with_scores_{MODEL_NAME}.csv")
    return df
```

```python
def display_results(model_name):
    df = return_df(model_name)
    
    top5, bottom5 = top_bottom(df, k=5)

    print("Top 5 abstracts:")
    for i in top5.abstract:
        print(i)

    print("\nBottom 5 abstracts:")
    for i in bottom5.abstract:
        print(i)

    print("Results for model:", model_name)
    print(df.alignment_score.describe())
```

```python
MODEL_NAMES = json.load(open("model_names.json"))
```

```python
display_results(MODEL_NAMES[0])
```

```python
display_results(MODEL_NAMES[1])
```

```python
display_results(MODEL_NAMES[2])
```
